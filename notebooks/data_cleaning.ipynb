Remove Technical Duplicates

These are rows that are identical in every single way. They usually happen due to system glitches during data export. You must remove these first to prevent data leakage.

```python
df = df.drop_duplicates(keep='first')

print(f"Shape after removing exact duplicates: {df.shape}")
```

```python
# Getting the entire dataset information
df.info()
```

Handling Data Types

```python
# Converting to datetime and extract features
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

```python
# Converting amount_src to Numeric
df['amount_src'] = (
    df['amount_src']
    .replace(r'[^\d.]', '', regex=True)
    .astype(float)
)
```

```python
# Ensuring booleans are 1/0 for the model
df['new_device'] = df['new_device'].astype(int)
df['location_mismatch'] = df['location_mismatch'].astype(int)
```

 Handle Timestamps

```python
# Filling missing timestamps with the median to avoid dropping data
df['timestamp'] = df['timestamp'].fillna(df['timestamp'].median())

df['hour_of_day'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```

Clean Numeric Fields

```python
# Impute missing amount_usd using amount_src * exchange_rate where possible
mask = df['amount_usd'].isnull()
df.loc[mask, 'amount_usd'] = df.loc[mask, 'amount_src'] * df.loc[mask, 
                                                                 'exchange_rate_src_to_dest']
```

```python
# Filling remaining nulls in fee and trust scores with medians
df['fee'] = df['fee'].fillna(df['fee'].median())
df['device_trust_score'] = df['device_trust_score'].fillna(df['device_trust_score'].median())
```

Categorical Columns

```python
# kyc
df['kyc_tier'].unique()
```

```python
# Standardizing casing and removing leading/trailing whitespace
df['kyc_tier'] = df['kyc_tier'].astype(str).str.strip().str.lower()

# Defining a clean mapping for misspellings and redundant labels
kyc_mapping = {
    'standrd': 'standard',
    'enhancd': 'enhanced',
    'nan': 'unknown',
    'none': 'unknown',
    'n/a': 'unknown'
}

# Applying the mapping
df['kyc_tier'] = df['kyc_tier'].replace(kyc_mapping)

# Handling actual NaN values
df['kyc_tier'] = df['kyc_tier'].replace('nan', 'unknown')
df['kyc_tier'] = df['kyc_tier'].fillna('unknown')

# Final validation
print(f"Cleaned KYC Tiers: {df['kyc_tier'].unique()}")
```

```python
# channel
df['channel'].unique()
```

```python
# Stripping whitespace and converting to lowercase
df['channel'] = df['channel'].str.strip().str.lower()

# Defining a mapping for misspellings and variations
channel_mapping = {
    'weeb': 'web',
    'mobille': 'mobile',
    'atm': 'atm',
    'unknown': 'other'
}

# Applying the mapping
df['channel'] = df['channel'].replace(channel_mapping)

# Verifying the fix
print(f"Cleaned Channels: {df['channel'].unique()}")
```

```python
# home country
df['home_country'].unique()
```

```python
# Casting to string, removing whitespace, and converting to uppercase for ISO consistency
df['home_country'] = df['home_country'].astype(str).str.strip().str.upper()

# Mapping all variants of 'nan' and 'none' to a standard 'UNKNOWN'
df['home_country'] = df['home_country'].replace(['NAN', 'NONE', 'NULL'], 'UNKNOWN')

# Final validation
print(f"Unique values in home_country: {df['home_country'].unique()}")
```

```python
# ip_country
df['ip_country'].unique()
```

```python
# String conversion, strip whitespace, and Uppercase
df['ip_country'] = df['ip_country'].astype(str).str.strip().str.upper()

# Consolidate Nulls: Map variants to 'UNKNOWN'
df['ip_country'] = df['ip_country'].replace(['NAN', 'NONE', 'NULL'], 'UNKNOWN')

# 3. Final Validation
print(f"Unique values in ip_country: {df['ip_country'].unique()}")
```
